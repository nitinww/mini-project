```python
import pandas as pd
```

```python
urls = [
    "https://www.moneycontrol.com/financials/w/ratiosVI/TM03/1",
    "https://www.moneycontrol.com/financials/w/ratiosVI/TM03/2",
    "https://www.moneycontrol.com/financials/w/ratiosVI/TM03/3",
    "https://www.moneycontrol.com/financials/w/ratiosVI/TM03/4"
]
```

```python
data = pd.DataFrame()
```

```python
for url in reversed(urls):
    df = pd.read_html(url)[0]
    df = df.iloc[:, :-1]
    df.columns = df.iloc[0]
    df = df.drop(0)
    df = df.reset_index(drop=True)
    df = df.drop(0)
    df.columns.values[0] = None
    df.set_index(df.columns[0], inplace=True)
    df = df.iloc[:, ::-1]
    data = pd.concat([data, df], axis=1)
```
